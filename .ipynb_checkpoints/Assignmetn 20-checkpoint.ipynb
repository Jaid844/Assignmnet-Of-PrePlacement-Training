```python
Given a binary tree, your task is to find subtree with maximum sum in tree.

Examples:

Input1 :       

       1

     /   \

   2      3

  / \    / \

4   5  6   7
```

```python
Construct the BST (Binary Search Tree) from its given level order traversal.

Example:

Input: arr[] = {7, 4, 12, 3, 6, 8, 1, 5, 10}

Output: BST:

            7

         /    \

       4     12

     /  \     /

    3   6  8

   /    /     \

 1    5      10
```

```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def constructBST(level_order):
    if not level_order:
        return None
    
    root = TreeNode(level_order[0])
    queue = [root]
    i = 1
    
    while i < len(level_order):
        current = queue.pop(0)
        
        if i < len(level_order):
            left_val = level_order[i]
            if left_val is not None:
                current.left = TreeNode(left_val)
                queue.append(current.left)
            i += 1
        
        if i < len(level_order):
            right_val = level_order[i]
            if right_val is not None:
                current.right = TreeNode(right_val)
                queue.append(current.right)
            i += 1
    
    return root

def inorderTraversal(node):
    if node is None:
        return []
    return inorderTraversal(node.left) + [node.val] + inorderTraversal(node.right)

# Example usage:
level_order = [7, 4, 12, 3, 6, 8, 1, 5, 10]

# Construct the BST from the level order traversal
root = constructBST(level_order)

# Perform inorder traversal to verify the constructed BST
inorder = inorderTraversal(root)
print("Inorder traversal of the constructed BST:", inorder)
```

```python
Given an array of size n. The problem is to check whether the given array can represent the level order traversal of a Binary Search Tree or not.

Examples:

Input1 : arr[] = {7, 4, 12, 3, 6, 8, 1, 5, 10}

Output1 : Yes

For the given arr[], the Binary Search Tree is:

            7

         /    \

       4     12

     /  \     /

    3   6  8

   /    /     \

 1    5      10
```

```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def isValidLevelOrder(arr):
    if not arr:
        return True
    
    nodeQueue = [TreeNode(arr[0])]
    rangeQueue = [(float('-inf'), float('inf'))]
    i = 1
    
    while nodeQueue:
        current = nodeQueue.pop(0)
        minVal, maxVal = rangeQueue.pop(0)
        
        if current.val <= minVal or current.val >= maxVal:
            return False
        
        if i < len(arr) and minVal < arr[i] < current.val:
            nodeQueue.append(TreeNode(arr[i]))
            rangeQueue.append((minVal, current.val))
            i += 1
        
        if i < len(arr) and current.val < arr[i] < maxVal:
            nodeQueue.append(TreeNode(arr[i]))
            rangeQueue.append((current.val, maxVal))
            i += 1
    
    return True

# Example usage:
arr = [7, 4, 12, 3, 6, 8, 1, 5, 10]

# Check if the array can represent a BST
is_valid = isValidLevelOrder(arr)
if is_valid:
    print("Yes, the array can represent a Binary Search Tree.")
else:
    print("No, the array cannot represent a Binary Search Tree.")
```
