```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X=np.random.rand(100,1)-0.5
```

```python
y=3*X[:,0]**2 + 0.05 * np.random.randn(100)
#data in specific manner
y
```

```python
import pandas as pd

df=pd.DataFrame()
```

```python
df['X']=X.reshape(100)

#single dimension data with 100 rows and -1 column

df['y']=y
```

```python
plt.scatter(df['X'], df['y'])
plt.title("X vs y")
```

```python
df.head()
```

```python
df['pred1']=df['y'].mean()

#Pred1 is base model ie. M0

df.head()
```

```python
df['res1']=df['y']=df['pred1']
df.head()
```

```python
plt.scatter(df["X"], df["y"])
plt.plot(df["X"], df["pred1"], color="red")

#first model prediction
```

```python

#Take decision tree on top of the residual column
```

```python

from sklearn.tree import DecisionTreeRegressor

tree1=DecisionTreeRegressor(max_leaf_nodes=8) 

#Threshold=8
```

```python
tree1.fit(df["X"].values.reshape(100,1),df["res1"].values)

#values to pass an array inside my fit

#use res1 to train my model
```

```python
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree

fig, ax = plt.subplots(figsize=(10, 10))
plot_tree(tree1, filled=True)
plt.show()
```

```python
X_test=np.linspace(-0.5, 0.5, 500)
X_test
```

```python
y_pred=tree1.predict(X_test.reshape(500,1))

#prediction

y_pred
```


#validating our model on unseen data.

```python

plt.figure(figsize=(14,4))
plt.subplot(121)
plt.plot(X_test, y_pred, linewidth=2, color="red")
plt.scatter(df["X"], df["y"])
```
